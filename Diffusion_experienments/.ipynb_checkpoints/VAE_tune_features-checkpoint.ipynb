```python
z = 0
for i in range(train_data.shape[0]):
    z += encoder(train_data[i]).detach()
z = z/train_data.shape[0]
```

```python
z.requires_grad_(True)
```

```python
def optimize_z(decoder, classifier, optimizer_z, criterion, target):
    decoder.eval()
    classifier.eval()
    optimizer_z.zero_grad()
    result = classifier(decoder(z))
    loss = criterion(result,attack_targets[target])
    loss.backward()
    optimizer_z.step()
    return loss.item()
    
```

```python
def optimize_z_eps(eps, lr, reg, lr_decay, target=8):
    for epoch in range(eps): 
        optimizer_z = optim.SGD([z], lr=lr*lr_decay**epoch, momentum=0.9, weight_decay=reg, nesterov=False)
        criterion = nn.CrossEntropyLoss(reduction = 'sum')
        epoch_loss = optimize_z(decoder, classifier, optimizer_z, criterion, target)
        print("loss(generator): ", epoch_loss)
        if epoch%10 == 0:
            tensor_to_picture(decoder(z).detach().cpu().numpy()[0][0])
```

```python
optimize_z_eps(10, 0.001, 0.000, 1, target=8)
```
