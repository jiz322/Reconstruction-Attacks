```python
import torch.optim as optim
from torch import nn

from preprocessing_last20 import *
from classifier import FaceRecognizer
from classifier2 import FaceRecognizer2
from mindiffusion.ddpm import DDPM
from mindiffusion.unet import NaiveUnet
```

```python
from torchvision import models

device = torch.device('cuda:0')
classifier = FaceRecognizer().to(device)
vgg11 = models.vgg11(pretrained=True).to(device)
classifier_eval = FaceRecognizer2(vgg11).to(device)
ddpm = DDPM(eps_model=NaiveUnet(1, 1, n_feat=128), betas=(1e-4, 0.02), n_T=600).to(device)
classifier.load_state_dict(torch.load('classifier.pt'))
classifier_eval.load_state_dict(torch.load('eval_classifier_overfit.pt'))
ddpm.load_state_dict(torch.load('ddpm_batched_600_last20.pth'))
```

```python
def sample_ddpm(noise=None):
    ddpm.eval()
    for param in ddpm.parameters():
        param.requires_grad = False
    torch.random.manual_seed(0)
    xh = ddpm.sample(1, (1, 64, 64), device, noise)
    tensor_to_picture(xh[0].detach().cpu().numpy()[0])
    return xh
```

```python
t7 = torch.load('attack7_lr1_10eps_overfit.pt').to(device).detach()
ddpm_attack7_10 = sample_ddpm(t7)
t7 = torch.load('attack7_lr1_20eps_overfit.pt').to(device).detach()
ddpm_attack7_20 = sample_ddpm(t7)
t7 = torch.load('attack7_lr1_30eps_overfit.pt').to(device).detach()
ddpm_attack7_30 = sample_ddpm(t7)
t7 = torch.load('attack7_lr1_40eps_overfit.pt').to(device).detach()
ddpm_attack7_40 = sample_ddpm(t7)
t7 = torch.load('attack7_lr1_50eps_overfit.pt').to(device).detach()
ddpm_attack7_50 = sample_ddpm(t7)
```

```python
print(torch.softmax(classifier_eval(ddpm_attack7_10), -1)[0][7])
print(torch.softmax(classifier(ddpm_attack7_10), -1)[0][7])
print(torch.softmax(classifier_eval(ddpm_attack7_20), -1)[0][7])
print(torch.softmax(classifier(ddpm_attack7_20), -1)[0][7])
print(torch.softmax(classifier_eval(ddpm_attack7_30), -1)[0][7])
print(torch.softmax(classifier(ddpm_attack7_30), -1)[0][7])
print(torch.softmax(classifier_eval(ddpm_attack7_40), -1)[0][7])
print(torch.softmax(classifier(ddpm_attack7_40), -1)[0][7])
print(torch.softmax(classifier_eval(ddpm_attack7_50), -1)[0][7])
print(torch.softmax(classifier(ddpm_attack7_50), -1)[0][7])
```
